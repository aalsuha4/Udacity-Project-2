### Droping  ScheduledDay, AppointmentDay columns and change No-show column name and datatype

Drop ScheduledDay and AppointmentDay columns and make sure they are droped

```python
# Droping ScheduledDay, AppointmentDay columns 
df.drop(['ScheduledDay', 'AppointmentDay'], axis=1,inplace=True)

# display the first row in the data set
df.head(1)
```

changing No-show column name to show because it would be easier to understand that 1 means show and 0 means did not show.

```python
# changing No-show column name to show 
df.rename(columns={'No-show': 'show'}, inplace=True)

# display the first row in the data set
df.head(1)
```

Replacing No to 1 and Yes to 0 and save it to the show colum.

```python
# replacing No to 1 and save it to the show colum
df.show = df.show.replace('No', '1')

# replacing Yes to 0 and save it to the show colum
df.show = df.show.replace('Yes', '0')

# display the first row in the data set
df.head(1)
```

After we change them to 0 and 1 we need to convert the colum to integer to make the colum more useful. 

```python
# converting show colum to integer
df.show = pd.to_numeric(df["show"])
```

Call head() to make sure that the changes have been made.

```python
# display the first 10 rows in the data set

df.head(10)
```

Age can't be less than 0 so we need to execute a query the get all the ages that are less than 0 and drop any.

```python
# execute a query the get all the ages that are less than 0
df.query('Age < 0')
```

```python
# drop all the ages that are less than 0 and update the dataset

df = df.drop(df[df['Age']<0].index)
```

After we made all these changes it would be better to save the modified data set to a new cvs file.

```python
# save the dataset in a new file with no index
df.to_csv('edited.csv', index=False)
```

We need to load the data set from new data set file to make sure every things look good.
